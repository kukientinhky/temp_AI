```python
# convert mp3 to wav
import os
import sys
import subprocess

def convert_mp3_to_wav(mp3_file, wav_file):
    command = "ffmpeg -i {0} -acodec pcm_s16le -ac 1 -ar 16000 {1}".format(mp3_file, wav_file)
    subprocess.call(command, shell=True)
    
for i in range(1, 9):
    mp3_file = "audio_input/{}.mp3".format(i)
    wav_file = "audio_input/0000{}.wav".format(i)
    convert_mp3_to_wav(mp3_file, wav_file)
    print("Converted {} to {}".format(mp3_file, wav_file))
```
